# Assignment Operators in Python

This notebook explains Python Assignment Operators with examples.

## Topics Covered
- Assignment (`=`)
- Add and Assign (`+=`)
- Subtract and Assign (`-=`)
- Multiply and Assign (`*=`)
- Divide and Assign (`/=`)
- Modulus and Assign (`%=`)
- Power and Assign (`**=`)

## Basic Assignment Operator

```
# Assign a value to a variable
x=10
print('Value of x:',x)
```

## Assignment Operators Examples

```
# Initial value
x=20

# Add and assign
x+=5
print('After += :',x)

# Subtract and assign
x-=3
print('After -= :',x)

# Multiply and assign
x*=2
print('After *= :',x)

# Divide and assign
x/=4
print('After /= :',x)

# Modulus and assign
x%=3
print('After %= :',x)

# Power and assign
x**=2
print('After **= :',x)
```

## Note
Assignment operators are used to update variable values quickly.
