# Relational Operators in Python

This notebook explores Python relational operators with examples.

## Topics Covered
- Equal to (`==`)
- Not equal to (`!=`)
- Greater than (`>`)
- Less than (`<`)
- Greater than or equal to (`>=`)
- Less than or equal to (`<=`)

```python
# Relational operators example
# Compare values and return True or False

a = 10
b = 5

print(a > b)
```

```python
# Relational operators example
# Compare values and return True or False

greater = a > b
greater_equal = a >= b


print(greater)
print(greater_equal)
```

```python
# Relational operators example
# Compare values and return True or False

a = 12
b = 10

less = a < b
less_equal = a <= b


print(less)
print(less_equal)
```

```python
# Relational operators example
# Compare values and return True or False

a = 15
b = 10

equal_to = a==b
not_equal = a != b

print(equal_to)
print(not_equal)
```
