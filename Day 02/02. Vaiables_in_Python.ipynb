 ## **Variables**

### Rules
* English Letters A-Z (Upper Case)
* English Letters a-z (Lower Case)
* Digits 0-9
* (_)
* Variables start from letters don't start to digit
* meaningful variable

```python
# Creating variables and storing values
name = "Muhammad Umair Khan"
age = 17
PI = 3.14

print(name,age,PI)
```

```python
# Printing the age variable
print("My age is :",age)
```

```python
# Performing a simple calculation with variables
print("My age is :",age-2)
```

```python
# Example of a valid variable name using underscore
_name = "Muhammad Umair Khan"

print(_name)
```

```python
# Example of an invalid variable name (cannot start with a digit)
4name = "Muhammad Umair Khan"

print(4name)
```
