# Python Data Types

This notebook explores basic Python data types using simple examples.

Topics covered:
- Integer
- Float
- String
- None Type


## Integer Data Type

Integers are whole numbers without decimal values.

Example:
- 10
- 100
- -5


```python
# Integer Data Type
# Store a whole number in a variable

Age = 13

# Print the data type of the variable
print(type(Age))
```

## Float Data Type

Float values contain decimal numbers.

Example:
- 3.14
- 2.5
- 0.99


```python
# Float Data Type
# Float stores decimal numbers

CGPA = 3.43

# Print the data type of the variable
print(type(CGPA))
```

## String Data Type

Strings are used to store text.

You can create strings with:
- Single quotes `' '`
- Double quotes `" "`
- Triple quotes `''' '''`


```python
# String Data Type
# Strings store text values

First_Name = 'Muhammad Umair'
Last_Name = "Khan"
Full_Name = '''Muhammad Umair Khan'''

# Print the data type of all string variables
print(type(First_Name))
print(type(Last_Name))
print(type(Full_Name))
```

## None Data Type

`None` means no value is assigned.

It is commonly used as a placeholder value in Python.


```python
# None Data Type
# None represents the absence of a value

Ali_Rupees = 500
Sharjeel_Rupees = None

# Print the data type of None variable
print(type(Sharjeel_Rupees))
```
