# Arithmetic Operators in Python

This notebook explores Python arithmetic operators with simple examples.

## Topics Covered
- Addition (`+`)
- Subtraction (`-`)
- Multiplication (`*`)
- Division (`/`)
- Modulus (`%`)
- Exponent (`**`)


```python
# Basic arithmetic operations

# Assign values to variables
a = 12
b = 13

# Addition
print(a+b)

# Subtraction
print(a-b)

# Multiplication
print(a*b)

# Division
print(a/b)
```

## Arithmetic Operations Example

Python provides arithmetic operators to perform mathematical calculations.


```python
# Store arithmetic operation results

a = 10
b = 5

# Addition
sum = a+b

# Subtraction
sub = a-b

# Multiplication
mult = a*b

# Division
divide = a/b

# Modulus (remainder)
modulo = a%b

# Power
power = a**b

# Print all results
print(sum)
print(sub)
print(mult)
print(divide)
print(modulo)
print(power)
```
