# Logical Operators in Python

This notebook explains Python Logical Operators with examples.

## Topics Covered
- AND (`and`)
- OR (`or`)
- NOT (`not`)

## AND Operator

```
# The AND operator returns True if both conditions are True
age=20
has_id=True
print(age>=18 and has_id)
```

## OR Operator

```
# The OR operator returns True if at least one condition is True
marks=45
sports_quota=True
print(marks>=50 or sports_quota)
```

## NOT Operator

```
# The NOT operator reverses the result
is_logged_in=False
print(not is_logged_in)
```

## Note

Logical operators are used to combine conditions and return True or False.
