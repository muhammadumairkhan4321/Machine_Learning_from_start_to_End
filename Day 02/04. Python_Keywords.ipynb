# Python Keywords

This notebook explains Python keywords and is GitHub-ready.

Keywords are reserved words with predefined meanings in Python.

## Python Keywords List

```
keywords=['False', 'None', 'True', 'and', 'as', 'assert', 'async', 'await', 'break', 'class', 'def', 'del', 'elif', 'else', 'except', 'finally', 'for', 'from', 'global', 'if', 'in', 'is', 'lambda', 'match', 'nonlocal', 'not', 'or', 'pass', 'raise', 'return', 'while', 'with', 'yield', 'case']

for i,k in enumerate(keywords,1):
    print(f'{i}. {k}')
```

## Note

- Keywords cannot be used as variable names.
- Keywords have special meanings in Python.
- Examples: `if`, `for`, `while`, `class`, `def`
